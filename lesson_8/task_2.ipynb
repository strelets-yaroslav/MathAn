__2.__ Исследовать на условный экстремум функцию

### $$U=2x^2+12xy+32y^2+15,$$

если

### $$x^2+16y^2=64$$

Функция Лагранжа:

$$L(x,y,\lambda) = 2x^2 + 12xy + 32y^2 + 15 + \lambda \cdot (x^2 + 16y^2 - 64)$$

Система уравнений частных производных первого порядка:

$$\begin{cases}
   L'_{x} = 4x + 12y + \lambda \cdot2x = 0, \\
   L'_{y} = 12x + 64y + \lambda \cdot32y = 0, \\
   L'_{\lambda} = x^2 + 16y^2 - 64 = 0
 \end{cases}$$
 
Выразим из второго уравнения $y$, и подставим в первое:
$$\begin{cases}
   4x - \frac{12*3x}{8(2 + \lambda)} + 2\lambda x = 0, \\
   y = -\frac{3x}{8(2 + \lambda)}, \\
   x^2 + 16y^2 - 64 = 0
 \end{cases}\Rightarrow 
 \begin{cases}
   \lambda^2 + 4\lambda + 7 = 0, \\
   y = -\frac{3x}{8(2 + \lambda)}, \\
   x^2 + 16y^2 - 64 = 0
 \end{cases}
 \Rightarrow 
 \begin{cases}
   \lambda = (-\frac{7}{2}, -\frac{1}{2}), \\
   y = -\frac{3x}{8(2 + \lambda)}, \\
   x^2 + 16y^2 - 64 = 0
 \end{cases}$$
 
Значения $x$ и $y$, удовлетворяющие этим условиям, равны: $x = \pm 4\sqrt{2},\,\,y = \pm \sqrt{2}$ (есть другие значения х и у для таких $\lambda$, но они не являются решениями изначальной системы уравнений).

При $\lambda = -\frac{7}{2}$ получаем две точки: $(4\sqrt{2}, \sqrt{2}, -\frac{7}{2})$ и $(-4\sqrt{2}, -\sqrt{2}, -\frac{7}{2})$

При $\lambda = -\frac{1}{2}$ получаем две точки: $(4\sqrt{2}, -\sqrt{2}, -\frac{1}{2})$ и $(-4\sqrt{2}, \sqrt{2}, -\frac{1}{2})$
 
Вторые производные:
$$L''_{xx}= 4 + 2\lambda$$
$$L''_{yy}= 64 + 32\lambda$$
$$L''_{\lambda \lambda}=0$$
$$L''_{xy}=L''_{yx}= 12$$
$$L''_{x\lambda}=L''_{\lambda x}= 2x$$
$$L''_{y\lambda}=L''_{\lambda y}= 32y$$ 

Матрица Гёссе:
$$\begin{pmatrix}
0 & 2x & 32y\\ 
2x & 4 + 2\lambda & 12\\
32y & 12 & 64 + 32\lambda\\
\end{pmatrix}$$

Определитель в общем виде:

$$\Delta=\begin{vmatrix}
0 & 2x & 32y\\ 
2x & 4 + 2\lambda & 12\\
32y & 12 & 64 + 32\lambda\\
\end{vmatrix}=-2x\cdot
\begin{vmatrix}
2x & 12\\
32y & 64 + 32\lambda\\
\end{vmatrix} + 2y\cdot
\begin{vmatrix}
2x & 4 + 2\lambda\\
32y & 12\\
\end{vmatrix} = -2x(2x(64 + 32\lambda) - 12\cdot32y) + 32y(2\cdot12x - 32y(4 + 2\lambda)) = 4\cdot64(6xy - 32\lambda - 64)$$

При $\lambda  = -\frac{1}{2}\,\, \Delta < 0$, т.е. $(4\sqrt{2}, -\sqrt{2}, -\frac{1}{2})$ и $(-4\sqrt{2}, \sqrt{2}, -\frac{1}{2})$ - точки минимума.

При $\lambda  = -\frac{7}{2}\,\, \Delta > 0$, т.е. $(4\sqrt{2}, \sqrt{2}, -\frac{7}{2})$ и $(-4\sqrt{2}, -\sqrt{2}, -\frac{7}{2})$ - точки максимума.

```python
import pylab
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.patches import Circle
import numpy as np
import warnings
warnings.filterwarnings('ignore')

fig = pylab.figure(figsize=(6,6))
axes = Axes3D(fig)


def makeData1 ():
    x = np.arange (-11, 12, 1)
    U = np.arange (4000, 5000, 5)
    xgrid, Ugrid = np.meshgrid(x, U)
    ygrid = np.sqrt(64 - xgrid**2)/4
    return xgrid, ygrid, Ugrid
x, y, U = makeData1()
axes.plot_surface(x, y, U, color='g', alpha=0.8)


def makeData2 ():
    x = np.arange (-11, 12, 1)
    U = np.arange (4000, 5000, 5)
    xgrid, Ugrid = np.meshgrid(x, U)
    ygrid = -np.sqrt(64 - xgrid**2)/4
    return xgrid, ygrid, Ugrid
x, y, U = makeData2()
axes.plot_surface(x, y, U, color='g', alpha=0.8)

def makeData3 ():
    x = np.arange (-10, 8, 1)
    y = np.arange (-10, 8, 1)
    xgrid, ygrid = np.meshgrid(x, y)
    Ugrid = 2*xgrid**2 + 12*xgrid*ygrid + 32*ygrid**2 + 15
    return xgrid, ygrid, Ugrid
x, y, U = makeData3()
axes.plot_surface(x, y, U, color='lightblue', alpha=0.6)

x1 = 4*np.sqrt(2)
y1 = np.sqrt(2)
z1 = 2*x1**2 + 12*x1*y1 + 32*y1**2 + 15

axes.scatter(x1, y1, z1, color='k')
axes.scatter(-x1, -y1, z1, color='k')

x2 = 4*np.sqrt(2)
y2 = -np.sqrt(2)
z2 = 2*x2**2 + 12*x2*y2 + 32*y2**2 + 15

axes.scatter(x2, y2, z2, color='r')
axes.scatter(-x2, -y2, z2, color='r')

pylab.show()

(z1, z2)
```
