__6*.__ Разложить функцию $y=e^x$ в ряд Маклорена, а так же в ряд Фурье на отрезке $[-\pi, \pi]$. Построить график функции и разложений.

- ряд Маклорена

$$f(x)=\sum\limits_{n=0}^{+\infty}\frac{f^{(n)}(0)}{n!}x^n = \sum\limits_{n=0}^{+\infty}\frac{x^n}{n!}$$

- ряд Фурье

$$f(x)=a_{0}+\sum\limits_{n=1}^{+\infty}(a_n\cos nx+b_n\sin nx)$$

где

$$a_0=\frac{1}{2\pi}\int\limits_{-\pi}^\pi f(x)\,dx$$

$$a_n=\frac{1}{\pi}\int\limits_{-\pi}^\pi f(x)\cos nx\,dx$$

$$b_n=\frac{1}{\pi}\int\limits_{-\pi}^\pi f(x)\sin nx\,dx$$

$$a_0=\frac{1}{2\pi}\int\limits_{-\pi}^\pi e^x\,dx = \frac{1}{2\pi}(e^{\pi} - e^{-\pi})$$

$$a_n = \frac{1}{\pi}\int\limits_{-\pi}^\pi e^x \cos nx\,dx$$

Берём по частям:

$$U=e^x\,\,\,\,\,\Rightarrow\,\,\,\,\, dU=e^x\,dx$$ 
$$dV=\cos nx\,dx\,\,\,\,\,\Rightarrow\,\,\,\,\, V=\frac{1}{n}\sin nx$$ 

$$=\Bigr(\frac{1}{\pi n}e^x\sin nx\Bigl)\Bigr|_{-\pi}^\pi - \frac{1}{\pi n}\int\limits_{-\pi}^\pi e^x\sin nx\,dx$$

Скобка равна нулю, т.к. $\sin \pi n=\sin (-\pi n)=0$. Интеграл возьмем еще раз по частям:

$$U=e^x\,\,\,\,\,\Rightarrow\,\,\,\,\, dU=e^x\,dx$$ 
$$dV=\sin nx\,dx\,\,\,\,\,\Rightarrow\,\,\,\,\, V=-\frac{1}{n}\cos nx$$ 

$$=\Bigr(\frac{1}{\pi n^2}e^x\cos nx\Bigl)\Bigr|_{-\pi}^\pi - \frac{1}{\pi n^2}\int\limits_{-\pi}^\pi e^x\cos nx\,dx$$

Переносим интеграл с косинусом влево и получаем:

$$a_n = \frac{(-1)^n(e^{2\pi} - 1)}{\pi e^{\pi}(n^2 + 1)}$$

$$b_n=\frac{1}{\pi}\int\limits_{-\pi}^\pi e^x\sin nx\,dx$$

Берём по частям:

$$U=e^x\,\,\,\,\,\Rightarrow\,\,\,\,\, dU=e^x\,dx$$ 
$$dV=\sin nx\,dx\,\,\,\,\,\Rightarrow\,\,\,\,\, V=-\frac{1}{n}\cos nx$$ 

$$=\Bigr(-\frac{1}{\pi n}e^x\cos nx\Bigl)\Bigr|_{-\pi}^\pi + \frac{1}{\pi n}\int\limits_{-\pi}^\pi e^x\cos nx\,dx$$

Интеграл возьмем еще раз по частям:

$$U=e^x\,\,\,\,\,\Rightarrow\,\,\,\,\, dU=e^x\,dx$$ 
$$dV=\cos nx\,dx\,\,\,\,\,\Rightarrow\,\,\,\,\, V=\frac{1}{n}\sin nx$$ 

$$=\Bigr(-\frac{1}{\pi n}e^x\cos nx\Bigl)\Bigr|_{-\pi}^\pi + \Bigr(\frac{1}{\pi n^2}e^x\sin nx\Bigl)\Bigr|_{-\pi}^\pi - \frac{1}{\pi n^2}\int\limits_{-\pi}^\pi e^x\sin nx\,dx$$

Скобка с $sin nx$ равна 0, т.к. $\sin \pi n=\sin (-\pi n)=0$. Переносим интеграл с синусом влево и получаем:

$$b_n = \frac{(-1)^n(1 - e^{2\pi})n}{\pi e^{\pi}(n^2 + 1)}$$

$$f(x)= \frac{1}{2\pi}(e^{\pi} - e^{-\pi}) + \sum\limits_{n=1}^{+\infty}\Bigl(\frac{(-1)^n(e^{2\pi} - 1)}{e^{\pi}(n^2 + 1)}\cos nx+ \frac{(-1)^n(1 - e^{2\pi})n}{e^{\pi}(n^2 + 1)}\sin nx \Bigr)$$

$$f(x) = \frac{1}{2\pi}(e^{\pi} - e^{-\pi}) + \frac{1}{\pi e^{\pi}}\sum\limits_{n=1}^{+\infty}\frac{(-1)^n}{(n^2 + 1)}\Bigl((e^{2\pi} - 1)\cos nx + n(1 - e^{2\pi})\sin nx \Bigr)$$

```python
import math as m
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
%matplotlib inline


pi = m.pi


def mcloren(x):
    return sum([x**i / m.factorial(i) for i in range(150)])


def furier(x):
    e_pi = m.e**pi
    
    a0 = (e_pi - 1 / e_pi) / (2 * pi)
    
    return a0 + sum([((e_pi**2 - 1)*m.cos(i*x) + i*(1 - e_pi**2)*m.sin(i*x))*(-1)**i / (i**2 + 1) for i in range(1, 10000)]) / (pi * e_pi)


step = pi/50
x = np.arange(-pi, pi, step)

plt.figure(figsize=(9, 9))
plt.plot(x, m.e**x, 'k')
plt.plot(x, mcloren(x), 'y--')
plt.plot(x, [furier(p) for p in x], 'g.')

plt.show
```

```python
x = np.arange(-pi + step, pi, step) # убрал левую границу - тогда по Фурье не будет начального выброса

plt.figure(figsize=(9, 9))
plt.plot(x, m.e**x, 'k')
plt.plot(x, mcloren(x), 'y--')
plt.plot(x, [furier(p) for p in x], 'g.')

plt.show
```
