### $$U=e^{-\frac{x}{2}}\cdot(x^2+y^2)$$

Найдем частные производные, приравняем их к нулю и найдем стационарные точки:

$$\begin{cases}
   U'_{x} = \frac{4x - x^2 - y^2}{2e^{\frac{x}{2}}} = 0, \\
   U'_{y} = \frac{2y}{e^{\frac{x}{2}}} = 0 
 \end{cases}$$
 
 При $y = 0$ получаем, что $x = 0$ и $x = 4$, т.е. имеются две точки $(0, 0)$, $(4, 0)$.
 
 Найдем вторые производные и составим матрицу Гёссе в общем виде:
 
$$U''_{xx} = \frac{(4 - 2x)2e^{\frac{x}{2}} - (4x - x^2 - y^2)e^{\frac{x}{2}}}{4(e^{\frac{x}{2}})^2} = \frac{8(1 - x) + x^2 + y^2}{4e^{\frac{x}{2}}}$$ $$U''_{xy} = U''_{yx} = -\frac{y}{e^{\frac{x}{2}}}$$  $$U''_{yy} = \frac{2}{e^{\frac{x}{2}}}$$

$$\begin{pmatrix}
\frac{8(1 - x) + x^2 + y^2}{4e^{\frac{x}{2}}} & -\frac{y}{e^{\frac{x}{2}}}\\ 
-\frac{y}{e^{\frac{x}{2}}} & \frac{2}{e^{\frac{x}{2}}}\\
\end{pmatrix}$$

Проверяем знаки в каждой точке.

$$(0, 0)$$

$$\Delta_{1} = \frac{8(1 - 0) + 0^2 + 0^2}{4e^{\frac{0}{2}}} = 2 > 0$$

$$\Delta_{2} = \frac{8(1 - 0) + 0^2 + 0^2}{4e^{\frac{0}{2}}}*\frac{2}{e^{\frac{0}{2}}} - (-\frac{0}{e^{\frac{0}{2}}})^2 = 4 > 0$$

Т.е. $(0, 0)$ - точка минимума.

$$(4, 0)$$

$$\Delta_{1} = \frac{8(1 - 4) + 4^2 + 0^2)}{4e^{\frac{4}{2}}} = -\frac{2}{e^2} < 0$$

$$\Delta_{2} = \frac{8(1 - 4) + 4^2 + 0^2)}{4e^{\frac{4}{2}}}*\frac{2}{e^{\frac{4}{2}}} - (-\frac{0}{e^{\frac{4}{2}}})^2 = -\frac{4}{e^4} < 0$$

Т.е. $(4, 0)$ - седловая точка.

```python
import pylab
import warnings
warnings.filterwarnings('ignore')
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

def makeData (x_start, x_end, y_start, y_end):
    x = np.arange (x_start, x_end, 0.1)
    y = np.arange (y_start, y_end, 0.1)
    xgrid, ygrid = np.meshgrid(x, y)

    Ugrid = (xgrid**2 + ygrid**2) / (np.exp(x / 2))
    return xgrid, ygrid, Ugrid

x, y, U = makeData(-10, 10, -10, 10)

fig = pylab.figure()
axes = Axes3D(fig)

axes.plot_surface(x, y, U)

pylab.show()
```

```python
# (0,0)
x, y, U = makeData(-1, 1, -1, 1)

fig = pylab.figure()
axes = Axes3D(fig)

axes.plot_surface(x, y, U)

pylab.show()
```

```python
# (4, 0)
x, y, U = makeData(3, 5, -1, 1)

fig = pylab.figure()
axes = Axes3D(fig)

axes.plot_surface(x, y, U)

pylab.show()
```
