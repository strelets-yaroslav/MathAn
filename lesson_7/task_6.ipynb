__6.__ Исследовать на экстремум функцию:

### $$U=x^2y+\frac{1}{3}y^3+2x^2+3y^2-1$$

Найдем частные производные, приравняем их к нулю и найдем стационарные точки:

$$\begin{cases}
   U'_{x} = 2xy + 4x = 0, \\
   U'_{y} = x^2 + y^2 + 6y = 0 
 \end{cases}$$
 
Получаем $y = -6$ и $y = 0$ при $x = 0$, а при $y = -2$ получаем $x = \pm2\sqrt{2}$, т.е. мы имеем четыре точки: $(0, 0)$, $(0, -6)$, $(-2\sqrt{2}, -2)$ и $(2\sqrt{2}, -2)$.
 
Найдем вторые производные и составим матрицу Гёссе в общем виде:
 
$$U''_{xx} = 2y + 4$$ $$U''_{xy} = U''_{yx} = 2x$$  $$U''_{yy} = 2y + 6$$

$$\begin{pmatrix}
2y + 4 & 2x\\ 
2x & 2y + 6\\
\end{pmatrix}$$

Вычисляем главные миноры:

$$\Delta_{1}=U''_{xx} = 2y + 4$$

$$\Delta_{2}=\begin{vmatrix}
U''_{xx} & U''_{xy}\\ 
U''_{yx} & U''_{yy}\\
\end{vmatrix}=U''_{xx}\cdot U''_{yy}-(U''_{xy})^2 = 4(y+2)(y+3) - 4x^2$$

Проверяем знаки в каждой точке.

$$(0, 0)$$

$$\Delta_{1} = 2y + 4 = 4 > 0$$

$$\Delta_{2} = 4(0+2)(0+3) - 4*0^2 = 24 > 0$$

Т.е. $(0, 0)$ - точка минимума.

Полный дифференциал второго порядка: $dU^2 = U''_{xx}dx^2+2U''_{xy}dxdy+U''_{yy}dy^2 = (2y + 4)dx^2 + 4xdxdy + (2y + 6)dy^2 = 4dx^2 + 6dy^2$

При любых, отличных от нуля, $dx$ и $dy$ полный дифференциал второго порядка положительный, это так же говорит о точке минимума.

$$(0, -6)$$

$$\Delta_{1} = 2y + 4 = -8 < 0$$

$$\Delta_{2} = 4(-6+2)(-6+3) - 4*0^2 = 48 > 0$$

Т.е. $(0, -6)$ - точка максимума

Полный дифференциал второго порядка: $dU^2 = U''_{xx}dx^2+2U''_{xy}dxdy+U''_{yy}dy^2 = (2y + 4)dx^2 + 4xdxdy + (2y + 6)dy^2 = -8dx^2 - 6dy^2$

При любых, отличных от нуля, $dx$ и $dy$ полный дифференциал второго порядка отрицательный, это так же говорит о точке максимума.

$$(-2\sqrt{2}, -2)$$

$$\Delta_{1} = 2y + 4 = 0$$

$$\Delta_{2} = 4(-2+2)(-2+3) - 4*(-2\sqrt{2})^2 = 32 > 0$$

Т.е. надо применять метод с касательной плоскостью

$$(2\sqrt{2}, -2)$$

$$\Delta_{1} = 2y + 4 = 0$$

$$\Delta_{2} = 4(-2+2)(-2+3) - 4*(2\sqrt{2})^2 = 32 > 0$$

Т.е. надо применять метод с касательной плоскостью

```python
import pylab
import warnings
warnings.filterwarnings('ignore')
from mpl_toolkits.mplot3d import Axes3D
import numpy

def makeData (x_start, x_end, y_start, y_end):
    x = numpy.arange (x_start, x_end, 0.1)
    y = numpy.arange (y_start, y_end, 0.1)
    xgrid, ygrid = numpy.meshgrid(x, y)

    Ugrid = xgrid**2 * y + ygrid**3 / 3 + 2*xgrid**2 + 3*ygrid**2 - 1
    return xgrid, ygrid, Ugrid

x, y, U = makeData(-10, 10, -10, 10)

fig = pylab.figure()
axes = Axes3D(fig)

axes.plot_surface(x, y, U)

pylab.show()
```

```python
import numpy as np
np.set_printoptions(precision=2, suppress=True)

# точка (0, 0)
a = np.array([[4, 0], [0, 6]])
w, v = np.linalg.eig(a)

print(f'Матрица Гёссе:\n{a}')
print(f'Собственные значения:\n{w}')
```

```python
# около (0, 0)
x, y, U = makeData(-2, 2, -2, 2)

fig = pylab.figure()
axes = Axes3D(fig)

axes.plot_surface(x, y, U)

pylab.show()
```
