__7.__ Найти длину $𝑥$ и ширину $𝑦$ прямоугольника при заданном периметре $𝑃=144$ cм, при которых данный прямоугольник имеет наибольшую площадь $𝑆$.

Составим систему уравнений по данным:

$$
\left\{
    \begin{array}\\
        a + b = P / 2 = 72,\\
        a*b = S
    \end{array}
\right.
$$

После выражения b через a получим $S = a(72 - a) = 72a - a^2$.

Найдём производную: $S' = 72 - 2a$. Находим критические точки: $S' = 0 => 72 - 2a = 0, a = 36$.

Проверим окрестность этой точки: $S'(35) = 72 - 70 = 2 > 0, S'(37) = 72 - 74 = -2 < 0$. Производная меняет знак с "+" на "-", следовательно, данная точка является максимумом.

Находим b: $b = 72 - a = 36$.

Таким образом при ширине 36см и длине 36см полученный прямоугольник (конечная форма является квадратом) будет иметь наибольшую площадь.

```python
from matplotlib import pylab as plt
import numpy as np
%matplotlib inline

plt.figure(figsize=(6,6))

x = np.linspace(1, 71, 200)
y = 72*x - x**2
plt.plot(x, y, c='g')

plt.axvline(x=36)

plt.show()
```
