__8*.__ Завод $A$ отстоит от железной дороги, идущей с юга на север и проходящей через $B$, считая по кратчайшему расстоянию на $a$ км. Под каким углом $\alpha$ к железной дороге следует ппостроить подъезной  путь от завода, чтобы транспортировка грузов из $A$ в $B$ была наиболее экономичной, если стоимость провоза тонны груза на расстоянии $1$ км составляет: по подъезному пути $p$ рублей, по железной дороге $q$ рублей $(p>q)$ и город $B$ расположен на $b$ км севернее завода $A$.

Выразим расстояние по ж/д от подъездной дороги до города через $x$ - будем искать оптимальное расстояние по ж/д для минимальной стоимости. Тогда общая стоимость пути (C = Cost) будет равна $C = p*\sqrt{a^2 + (b - x)^2} + q*x$, т.к. расстояние от завода до дороги можно рассчитать по формуле для гипотенузы, где катеты - это $a$ и $b - x$.

Найдём производную: $C' = p*\frac{1}{\sqrt{a^2 + (b - x)^2}}*2(b - x)*(-1) + q = q - \frac{p(b - x)}{\sqrt{a^2 + (b - x)^2}}$

Найдём критические точки для $C' = 0$:

$$q - \frac{p(b - x)}{\sqrt{a^2 + (b - x)^2}} = 0$$
$$q*\sqrt{a^2 + (b - x)^2} - p(b - x) = 0$$
$$q^2*(a^2 + (b - x)^2) - p^2(b - x)^2 = 0$$
$$(b - x)^2(p^2 - q^2) = (qa)^2$$
$$(b - x) = \frac{qa}{\sqrt{p^2 - q^2}}$$
$$x = b - \frac{qa}{\sqrt{p^2 - q^2}}$$

$x$ входит в уравнение в нечётной степени (первой), тогда надо проверить знак производной справа - для этого возьмём $x = b$. Тогда $C'(b) = q - \frac{p*(b - b)}{\sqrt{a^2 + (b - b)^2}} = q > 0$, т.е. производная в критической точке меняет знак с "-" на "+" -> данная точка является точкой минимума.

Теперь вычислим угол для подъездной дороги: $\mbox{tg}(\alpha) = \frac{a}{b - x} = \frac{a}{\frac{qa}{\sqrt{p^2 - q^2}}} = \frac{\sqrt{p^2 - q^2}}{q}$, $\alpha = \mbox{arctg}\Bigl( \frac{\sqrt{p^2 - q^2}}{q} \Bigr)$

```python
from matplotlib import pylab as plt
import numpy as np
%matplotlib inline

plt.figure(figsize=(6,6))

a = 10
b = 30
p = 5
q = 3

x = np.linspace(0, 40, 200)
y = p*np.sqrt(a**2 + (b - x)**2) + q*x
plt.plot(x, y, c='g')

plt.axvline(x = (b - (q*a) / (np.sqrt(p**2 - q**2))))

plt.show()
```
